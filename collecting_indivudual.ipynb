# MoMA EDA

One of the most popular museums in the world is MoMA (Museum of Modern Art) in New York City, which has a sizable collection of contemporary art. 

MoMA makes a dataset of their collection available for download. The dataset contains information about artworks, artists, and other topics. The dataset is available for download from the MoMA website and can be used for analysis and study. This dataset has been released into the public domain under a CC0 licence.

```python
pd.set_option('display.max_columns', None)
```

```python
artworks = pd.read_csv("https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artworks.csv", delimiter=',', encoding='utf-8')

artworks
```

```python
artists = pd.read_csv('https://media.githubusercontent.com/media/MuseumofModernArt/collection/master/Artists.csv',delimiter=',', encoding='utf-8')

artists
```

The artworks dataset has 140,848 records that reflect all of the works that have been added to the MoMA collection, while the Artists dataset contains 15,243 records that represent all of the artists who have work in the MoMA collection.

## CLEAN DATA

In this section, I will merge two datasets and clean the data to make it more queryable.

```python
artists = artists.rename(columns={'DisplayName': 'Artist'})
artworks = artworks.drop(['ConstituentID','ArtistBio','Nationality','BeginDate','EndDate','Gender'],axis = 1)
df = pd.merge(artists, artworks, on='Artist')
df
```

```python
df = df.drop(['Wiki QID','URL','ThumbnailURL','ULAN','Cataloged'],axis = 1)
df.head()
```

```python
df.duplicated(keep=False)
```

```python
df.dtypes
```

### About Artists

To begin, I'd like to dig a little deeper to discover what the male-to-female ratio is.

```python
df.groupby('Gender').count()['ConstituentID']
```

Here, I changed the form of the data and searched up the "Non-binary" data to allow for a better analysis.

```python
df['Gender'] = df['Gender'].replace("male", "Male")
df['Gender'] = df['Gender'].replace("female", "Female")
df['Gender'] = df['Gender'].replace("Non-binary", "Non-Binary")
```

```python
non_binary_df = df[df['Gender'].str.contains("Non-Binary", na=False)]
non_binary_df
```

Following my search for these three artists, I believe they may have chosen not to identify their gender. Some people may prefer not to disclose this information, and as a result, it may not be included in the dataset. So I decided to leave things alone.

```python
df.groupby('Gender').count()['ConstituentID']
```

```python
df['Gender'].value_counts().plot(kind='bar', title= "MoMA Artists'gender distribution", color = ['blue','orange','purple'])
```

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.3f%%', title= "MoMA Gender Proportion Among Artists")
```

Here, we count the years in which the most artists were born and died.

```python
df = df[df['BeginDate'] != 0]
```

```python
df.groupby('BeginDate').count()['ConstituentID'].plot(kind='line',title='Number of Artists by Date of Birth')
```

```python
df = df[df['EndDate'] != 0]
df.groupby('EndDate').count()['ConstituentID'].plot(kind='line',title='Number of Artists by Date of Death')
```

It can be seen that many artists were born after 1850 and that there was a peak around 1850. And for the death, around 1930, the greatest number of people perished. At last, let's see the artists' age distribution.

```python
#First, I checked if the artist is alive or not
df['Alive'] = df['EndDate'].isna()

#Calculate the age of the alive artists
df.loc[df['Alive'] == True, 'Age'] = 2022 - df['BeginDate']

#Calculate the age of the deceased artists
df.loc[df['Alive'] == False, 'Age'] = df['EndDate'] - df['BeginDate']

#Plot the histogram
df['Age'].plot(kind='hist', bins=30, color='green', title='MOMA Artists Age Distribution')

plt.xlabel('Age')
plt.ylabel('Number of Artists')
plt.title('MOMA Artists Age Distribution')

plt.show()
```

Now I'd like to find out where most of the artists are from.


```python
nationality = df.Nationality.value_counts()
nationality
```

There are 82 different nationalities represented here (excluding works without nationality data)and because there are so many, I've selected the top 10 for a closer look.

```python
nationality.head(20)
```

As can be seen, most of the top twenty countries are European, with Japan representing Asia.

```python
top_10_nationalities = df['Nationality'].value_counts().nlargest(20)

top_10_nationalities.plot(kind='bar', color=['red','yellow','blue','orange','green','violet','indigo','purple','pink','gray'],title='MoMA Top 20 Nationality of Artists')

plt.xlabel('Nationality')
plt.ylabel('Number of Artists')
plt.title('MoMA Top 20 Nationality of Artists')

plt.show()
```

From the plot, we can see majority of the artists lived up to 60 years old. The data revealed that many artists lived rather long lives.

### About Artworks

MoMA has a variety of art collections, and we'll explore the artworks in the MOMA in this part, including their acquisition background. 

```python
creditLine_counts = df.CreditLine.value_counts()
creditLine_counts.head(10)
```

```python
df.CreditLine.value_counts()[:10].plot(kind = 'bar', title = 'MoMA Top 10 Collection Source')
```

We can see that the Louis E. Stern collection contributed the most. Many of the artworks are then gifted by different artists.

```python
classification_counts = df.Classification.value_counts()
classification_counts.head(10)
```

```python
# Create the time periods
time_periods = pd.cut(df['AcquisitionYear'], bins=10)
time_period_count = df.groupby(time_periods).count()['ConstituentID']

time_period_count.plot(kind='bar', title='MOMA Artwork Acquisition Year Distribution')
```

And the largest number of artworks were acquired between 1956 and 1966.

Then what is the largest classification of artworks acquired?

```python
df.Classification.value_counts()[:10].plot(kind = 'bar', title = 'Top 10 Used Medium')
```

We can see that the majority of the collection is made up of illustrated books and prints, as well as photographs.
